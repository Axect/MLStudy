<h2 style="text-align:center">Brief intro R</h2>

### Pros of R

---

* Open Source
* Statistics Library
* OOP & FP
* Visualization
* Optimized to Statistics
* Easy to Learn (Not for advanced developer!)

**For example,**

```python
# Python should import np
import numpy as np

p1 = np.array([i for i in range(1,11)])
p2 = p1 ** 2 + p1
print(p1)
```

```R
# R is more intuitive
x <- seq(1,10,1)
y <- x^2 + x
print(y)
summary(y) # summary is greate!
```
